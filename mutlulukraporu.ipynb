# WORLD HAPPINESS REPORT
---
* We have analyzed world happiness score dataset which consists of different features such as economic production, social support etc.
* In this study, we studied the most happy and the least happy 10 countries in 2015, 2016 and 2017.
* After that, we tried to identify the top 10 countries according to factors that contributes to happiness level. These factors were economy, family, life-expectancy, freedom, generosity and government corruption
* Finally, we calculated the difference in happiness score between 2015 and 2019. We sum up with the most increased happiness score for 10 countries and the most decreased happiness score for 10 countries. 
* Our data set is generated from kaggle : https://www.kaggle.com/datasets/unsdsn/world-happiness

```python
# we tried to concatenate all 5 data but we need to transform them to become similar
# we dropped all columns except country and happiness score
# we changed all column and row names in order to concatenate them
# at last, we concatenate all scores in horizontal direction
data2015=data2015.sort_values(by="Country").loc[:,["Country","Happiness Score"]].rename(columns={"Happiness Score":"Skore2015"}).set_index('Country')
data2016=data2016.sort_values(by="Country").loc[:,["Country","Happiness Score"]].rename(columns={"Happiness Score":"Skore2016"}).set_index('Country')
data2017=data2017.sort_values(by="Country").loc[:,["Country","Happiness.Score"]].rename(columns={"Happiness.Score":"Skore2017"}).set_index('Country')
data2018=data2018.sort_values(by="Country or region").loc[:,["Country or region","Score"]].rename(columns={"Country or region":"Country","Score":"Skore2018",}).set_index('Country')
data2019=data2019.sort_values(by="Country or region").loc[:,["Country or region","Score"]].rename(columns={"Country or region":"Country","Score":"Skore2019",}).set_index('Country')
new_data=pd.concat([data2015,data2016,data2017,data2018,data2019],axis=1)
```

```python
#check the null values and drop them
new_data.isnull().sum()
```

```python
#after dropping null values, we need to redefine the data
new_data.dropna(inplace=True)
```

```python
#review the data, just first 5 row
new_data.head()
```

```python
#the difference in happiness score between 2015 and 2019
new_data["Fark"]=new_data["Skore2019"]-new_data["Skore2015"]
```

```python
#sorting the new_data from the highest difference to lowest difference
#indexing new_data again and create a "Country" column which includes indexes
#presenting barplot concerning descending difference in countries
#rotating x-parameters which is "Country" for -45 degree
#to sum up, this barplot show us 10 countries which vary from the highest happiness difference to the lowest happiness difference
new_data=new_data.sort_values(by="Fark",ascending=False)
new_data["Country"]=new_data.index
ax=sns.barplot(data=new_data.head(10),x="Country",y="Fark")
ax.tick_params(axis = 'x', rotation = 45)
plt.title('Mutluluk skoru en ok artan lkeler')
plt.show()
```

```python
#Similar algorithms but this time, we start presenting from the highest decrease in happiness to the lowest one
#In order to create nice plot, we converted minus happiness score to plus happiness score via "abs"
new_data=new_data.sort_values(by="Fark")
new_data["Fark"]=abs(new_data["Fark"])
ax=sns.barplot(data=new_data.head(10),x="Country",y="Fark")
ax.tick_params(axis = 'x', rotation = 45)
plt.title('Mutluluk skoru en ok azalan lkeler')
plt.show()
```
